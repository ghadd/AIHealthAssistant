```python
df1 = pd.read_csv('./cleaned_data_ru.csv', error_bad_lines=False)
df1.head()
```

```python
df1.shape
```

```python
df1 = df1.drop(['0'], axis=True)
```

```python
df1.columns = ['disease', 'symptom', 'occurence_count']
```

```python
# Remove any rows with empty values
df1.replace(float('nan'), np.nan, inplace=True)
df1.dropna(inplace=True)
```

```python
from sklearn import preprocessing
```

```python
n_unique = len(df1['symptom'].unique())
n_unique
```

```python
df1.dtypes
```

```python
# Encode the Labels
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(df1['symptom'])
print(integer_encoded)
```

```python
# One Hot Encode the Labels
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
```

```python
onehot_encoded[0]
```

```python
len(onehot_encoded[0])
```

```python
cols = np.asarray(df1['symptom'].unique())
cols
```

```python
# Create a new dataframe to save OHE labels
df_ohe = pd.DataFrame(columns = cols)
df_ohe.head()
```

```python
for i in range(len(onehot_encoded)):
    df_ohe.loc[i] = onehot_encoded[i]
```

```python
df_ohe.head()
```

```python
len(df_ohe)
```

```python
# Disease Dataframe
df_disease = df1['disease']
df_disease.head()
```

```python
# Concatenate OHE Labels with the Disease Column
df_concat = pd.concat([df_disease,df_ohe], axis=1)
df_concat.head()
```

```python
df_concat.drop_duplicates(keep='first',inplace=True)
```

```python
df_concat.head()
```

```python
len(df_concat)
```

```python
cols = df_concat.columns
df_concat.to_csv("df_concat.csv")
cols
```

```python
cols = cols[1:]
```

```python
# Since, every disease has multiple symptoms, combine all symptoms per disease per row
df_concat = df_concat.groupby('disease').sum()
df_concat = df_concat.reset_index()
df_concat[:5]
```

```python
len(df_concat)
```

```python
df_concat.to_csv("./training_dataset.csv", index=False)
```

```python
# One Hot Encoded Features
X = df_concat[cols]

# Labels
y = df_concat['disease']
```

## Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
len(X_train), len(y_train)
```

```python
len(X_test), len(y_test)
```

```python
dt = DecisionTreeClassifier()
clf_dt=dt.fit(X, y)
```

```python
clf_dt.score(X, y)
```

```python
disease_pred = clf_dt.predict(X)
```

```python
disease_real = y.values
```

```python
for i in range(0, len(disease_real)):
    if disease_pred[i]!=disease_real[i]:
        print ('Pred: {0}\nActual: {1}\n'.format(disease_pred[i], disease_real[i]))
```

```python
import spacy    
nlp = spacy.load("ru_core_news_lg")
```

```python
def get_prediction(model, query : str):
    query = query.lower()
    query = nlp(query)
    
    query_vec = []
    for symp in df_concat.columns[1:]:
        query_vec.append(int(query.similarity(nlp(symp)) > 0.5))

    res = model.predict([
        np.array(query_vec)
    ])
    
    print(res)
```

```python
get_prediction(clf_dt, "Острая боль в голове")
```
