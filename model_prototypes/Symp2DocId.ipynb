```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./comments_doctor.csv", index_col=0)
df
```

```python

def drop_excess(s):
    s = s.replace("'", '').replace('\n', '').replace('[', '').replace(']', '')
    return s.split()
```

```python
df.comment = df.comment.apply(drop_excess)
```

```python
X = np.array([], dtype=np.str_)
y = np.array([], dtype=np.int16)


i = 0
for row in df.iterrows():
    print(f"\r{i * 100 / len(df)}%", end="")
    
    comment = row[1][0]
    doc_id = row[1][1]
    
    X = np.append(X, comment)
    y = np.append(y, np.full(len(comment), doc_id))
    
    i += 1
```

```python
df_train = pd.DataFrame({
    'X': X,
    'y': y
})
df_train
```

```python
df_train.shape
```

```python
df_train.to_csv("comment2doc_id.csv")
```

```python
# Encode the Labels
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(df_train.X)
print(integer_encoded)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(integer_encoded, y, test_size=0.2, random_state=101)
```

```python
dt = DecisionTreeClassifier()
clf_dt=dt.fit(X_train.reshape(len(X_train), 1), y_train)
```

```python
clf_dt.score(X, y)
```

```python
disease_pred = clf_dt.predict(X_test)
```
